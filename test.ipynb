```python
import json

with open('bbox-v1.0.json') as f:
    d = json.load(f)
```

```python
def get_bbox_value(vid, index, default=None):
    if vid in d:
        return d[vid][index]
    return default

# Create the new columns
final_df['x0'] = final_df['vid'].apply(lambda vid: get_bbox_value(vid, 0))
final_df['y0'] = final_df['vid'].apply(lambda vid: get_bbox_value(vid, 1))
final_df['x1'] = final_df['vid'].apply(lambda vid: get_bbox_value(vid, 2))
final_df['y1'] = final_df['vid'].apply(lambda vid: get_bbox_value(vid, 3))
```

```python
# Filter data_from_git to keep only rows with vid values found in final_df
filtered_data = data_from_git[data_from_git['vid'].isin(final_df['vid'])]

# Verify the results
print(f"Original data_from_git shape: {data_from_git.shape}")
print(f"Filtered data shape: {filtered_data.shape}")
print(f"final_df unique vid count: {final_df['vid'].nunique()}")
print(f"filtered_data unique vid count: {filtered_data['vid'].nunique()}")

# Check if all final_df vid values are represented in the filtered data
missing_vids = set(final_df['vid']) - set(filtered_data['vid'])
if missing_vids:
    print(f"Warning: {len(missing_vids)} vid values from final_df are missing in the filtered data")
else:
    print("All final_df vid values were found in data_from_git")
```
