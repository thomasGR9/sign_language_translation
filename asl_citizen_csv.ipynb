```python
import os

def delete_files_from_dataframe(df, path_column='file_path', confirm=True):
    """
    Delete all files specified in a dataframe column.
    
    Parameters:
    df (pandas.DataFrame): Dataframe containing file paths
    path_column (str): Name of the column containing file paths
    confirm (bool): Whether to ask for confirmation before deletion
    
    Returns:
    tuple: (deleted_count, failed_files) - Count of deleted files and list of files that couldn't be deleted
    """
    total_files = len(df)
    
    # Safety confirmation
    if confirm:
        confirm_input = input(f"WARNING: You are about to delete {total_files} files. Type 'yes' to confirm: ")
        if confirm_input.lower() != 'yes':
            print("Operation cancelled.")
            return 0, []
    
    deleted_count = 0
    failed_files = []
    
    # Iterate through all file paths and delete each file
    for file_path in df[path_column]:
        if not isinstance(file_path, str):
            failed_files.append(f"Invalid path (not a string): {file_path}")
            continue
            
        try:
            if os.path.exists(file_path):
                os.remove(file_path)
                deleted_count += 1
                print(f"Deleted: {file_path}")
            else:
                failed_files.append(f"File not found: {file_path}")
        except Exception as e:
            failed_files.append(f"Error deleting {file_path}: {str(e)}")
    
    print(f"Deleted {deleted_count} out of {total_files} files.")
    if failed_files:
        print(f"Failed to delete {len(failed_files)} files.")
    
    return deleted_count, failed_files

# Usage example:
# delete_files_from_dataframe(long_videos_df)
```

```python
import os
from tqdm import tqdm

def rename_videos_by_handedness(df, file_path_col='file_path', dom_hand_col='dom_hand', inplace=False):
    """
    Renames video files by adding a suffix (_R or _L) based on the dominant hand information.
    Updates the file paths in the dataframe.
    
    Parameters:
        df (pandas.DataFrame): Dataframe containing file paths and dominant hand info
        file_path_col (str): Name of the column containing file paths
        dom_hand_col (str): Name of the column containing dominant hand info ('Right' or 'Left')
        inplace (bool): Whether to modify the original dataframe or return a copy
        
    Returns:
        pandas.DataFrame: Updated dataframe with new file paths
    """
    # Use original dataframe or create a copy
    result_df = df if inplace else df.copy()
    
    # Initialize counters
    successful = 0
    failed = 0
    already_renamed = 0
    failed_files = []
    
    # Process each row
    for idx, row in tqdm(result_df.iterrows(), total=len(result_df), desc="Renaming files"):
        file_path = row[file_path_col]
        dom_hand = row[dom_hand_col]
        
        # Skip if file_path is not a string or dom_hand is missing
        if not isinstance(file_path, str) or pd.isna(dom_hand):
            failed += 1
            failed_files.append((file_path, "Invalid file path or missing dominant hand info"))
            continue
        
        # Skip if file doesn't exist
        if not os.path.exists(file_path):
            failed += 1
            failed_files.append((file_path, "File does not exist"))
            continue
        
        # Determine the suffix based on dominant hand
        if dom_hand == 'Right':
            suffix = "_R"
        elif dom_hand == 'Left':
            suffix = "_L"
        else:
            failed += 1
            failed_files.append((file_path, f"Unknown dominant hand value: {dom_hand}"))
            continue
        
        # Check if the file already has the suffix
        dir_name = os.path.dirname(file_path)
        file_name = os.path.basename(file_path)
        base_name, ext = os.path.splitext(file_name)
        
        if base_name.endswith('_R') or base_name.endswith('_L'):
            # File already has a handedness suffix
            already_renamed += 1
            continue
        
        # Create new file path with the suffix
        new_base_name = f"{base_name}{suffix}"
        new_file_name = f"{new_base_name}{ext}"
        new_file_path = os.path.join(dir_name, new_file_name)
        
        try:
            # Rename the file
            os.rename(file_path, new_file_path)
            
            # Update the dataframe
            result_df.at[idx, file_path_col] = new_file_path
            
            successful += 1
            
        except Exception as e:
            failed += 1
            failed_files.append((file_path, str(e)))
    
    # Report results
    print(f"\nRenaming complete:")
    print(f"  - Successfully renamed: {successful} files")
    print(f"  - Already had suffix: {already_renamed} files")
    print(f"  - Failed: {failed} files")
    
    if failed > 0:
        print("\nFailed files:")
        for path, error in failed_files:
            print(f"  - {os.path.basename(path) if isinstance(path, str) else path}: {error}")
    
    return result_df
```

```python
resampled_renamed_df = rename_videos_by_handedness(df=resampled_data, file_path_col='file_path', dom_hand_col='dom_hand', inplace=False)
```
